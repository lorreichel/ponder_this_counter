```python
from difflib import get_close_matches
```

```python
import json
json1_file = open('top_solvers_results.json')
json1_str = json1_file.read()
json1_data = json.loads(json1_str)
```

```python
names = [i['name'] for i in json1_data['all_time']]
```

```python
len(names)
```

```python
simlist = []
for name in names[:100]:
    # print(name, get_close_matches(name, names))
    simlist.append([name, get_close_matches(name, names)[1:]])
```

```python
for i in simlist[:20]:
    print(i)
```

```python
search_term = 'Radu'
for name in names:
    if search_term in name:
        print(name)
```

```python
alltimescores = {}
june24scores = {}
for s in json1_data['all_time']:
    alltimescores[s['name']] = s['count']
for s in json1_data['through_june_2024']:
    june24scores[s['name']] = s['count']
```

```python
for name in alltimescores:
    try:
        print(f'name: {name:<30},     total: {alltimescores[name]:>3} ,    j24: {june24scores[name]:>3},     inc={alltimescores[name]-june24scores[name]:>3}')
    except:
        print(name, 'no june 24 score')
```
